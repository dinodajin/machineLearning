### 문제 정의

- BX : 최근 이직 시장이 활발하게 성장함에 따라 직원들도 퇴사 행렬을 이루고 있습니다.. 경영진은 핵심인재 유출을 막기 위해 오랜시간 함께하고 싶은 회사를 만들고자 함.
- CX : 자신이 성장 가능한 회사, 성과에 따라 적절한 보상과 효율적인 업무를 진행할 수 있는 회사를 만들고자 함
- DX : 현재 직원들에 대한 이직 현황 데이터 분석을 통해 이직률과 연관 있는 데이터를 지속적으로 확인하고 개선하고자 한다. HR팀은 이직 가능성이 높은 핵심인재를 예측하고, 해당 인원들을 위한 관리 프로그램을 운영하고자 함
--------------------------------------------------------------------------------------------------------------------------------------------------------
- Trigger : HR 데이터를 통한 최근 퇴사 및 이직 현황을 확인, 핵심인재의 유출 여부를 확인
- Accelerator : 직원의 인사 관련 데이터를 통해 구성원의 불만족 요소 혹은 잠재적인 문제(보상, 진급, 가치관,..등)을 확인
- Tracker : 핵심 인재의 이직률 변화 추이, 직원들의 업무 스트레스 지수

```python
# 이직률 현황 확인
data['Attrition'].value_counts()
```

```python
# 이직률 계산
# 이직한 사람 / 전체 직원 수
237 / (1233 + 237)
# 약 16%
```

```python
# 이직 데이터 -> 학습과 집계를 위해 수치 데이터로 변경
# Yes -> 1, No -> 0
# np.where(구간을 나눌 조건, 조건이 참일 경우 넣을 값, 그 외 데이터) -> 2개의 구간으로 나눠주는 함수
data['Attrition'] = np.where(data['Attrition'] == 'Yes', 1, 0)
data['Attrition'].value_counts()
```

```python
# Age 연령데이터 확인
data['Age']
```

```python
# 나이 데이터 카테고리화(범주화) 하기 -> where, cut 이용가능
# 30대 미만, 30-40대, 40대 이상
data['Age_cate'] = np.where(data['Age'] < 30, '30대 미만', 
         np.where(data['Age'] < 40, '30-40대', '40대 이상'))
```

```python
pd.set_option("display.max_rows", None)
data[['Age','Age_cate']]
```

```python
# 나이대별 이직률 확인
age_gb = data.groupby('Age_cate')['Attrition'].agg(['count', 'sum'])
# count : 해당 카테고리에 있는 인원수 (데이터가 있으면 셈)
# sum : 해당 카테고리에 있는 이직자(1) 수 (1인 데이터만 더함)
# 즉, 30대 미만인 326명의 직원 중 91명이 이직함
age_gb
```

```python
# 나이대 별 이직률 현황 구하기
age_gb['ration'] = round(age_gb['sum'] / age_gb['count'] * 100, 1) # 소수점 첫 번째 자리까지만 출력
age_gb
```

```python
# 성별에 따른 이직률 구하기
data['Attrition'] = np.where(data['Attrition'] == 'Yes', 1, 0)
gender_gb = data.groupby('Gender')['Attrition'].agg(['count', 'sum'])
gender_gb
```

- 가설 1) 업무 만족도는 높으나 인간 관계로 인한 이직률이 높을 것이다!
- -> 가설이 옳지 않음, 반대임

```python
# 인간 관계에 대한 만족도 : RelationshipSatisfaction
# 업무 만족도 : JobSatisfaction
# 숫자가 클수록 만족도가 높다

data[['RelationshipSatisfaction', 'JobSatisfaction', 'Attrition']].head()
```

```python
# 업무 만족도 / 인간 관계의 상관관계를 통한 이직률 현황
gb = data.groupby(['JobSatisfaction', 'RelationshipSatisfaction'])['Attrition'].agg(['count', 'sum'])
gb['ratio'] = round(gb['sum'] / gb['count'] * 100, 1)
gb
# 결과1 : 업무 만족도가 높은 직원일수록 인간 관계에 대한 이직 영향을 덜 받는다
# 결과2 : 업무 만족도가 낮은 직원일수록 인간 관계에 대해 이직 영향을 많이 받는다
# 결과 : 전체적으로 인간 관계보다는 업무 만족도가 이직률에 더 영향을 끼친다
```

- 가설 2) 근속년수 대비 같은 업무를 한 비중이 높다면 이직률이 높을 것이다!
- -> 가설이 옳지 않음, 한 업무를 오래하도록 하는 것이 좋다

```python
# YearsInCurrentRole : 직원이 현재 역할에서 근무한 기간
# YearsAtCompany : 직원이 현재까지 근무한 기간
data[['YearsInCurrentRole', 'YearsAtCompany']]
```

```python
# 근속 년수 대비 한가지 업무를 한 비율
data['Role_Company'] = data['YearsInCurrentRole'] / data['YearsAtCompany']
data['Role_Company'] # 결측치 발생 -> 결측치와 결측치의 연산 결과 -> 결측치
data['Role_Company'].fillna(0, inplace = True)
# fillna(채우고 싶은 값) : 결측치를 내가 원하는 값으로 채울 때 사용하는 함수로서, inplace 속성을 적용해야만 저장 ( inplace 속성 확인)
```

```python
# 경고창 무시
import warnings
warnings.filterwarnings('ignore')
```

```python
# data['Role_Company'] -> 분포 확인 0> distplot
sns.distplot(data['Role_Company'])
```

```python
# Role_Company가 0인 데이터의 수치 확인
len(data[data['Role_Company'] == 0])
# data['Role_Company'] == 0인 244명에 대한 데이터는 제외하고 구간화를 진행하자
```

```python
# 0은 제외하고 구간화 작업 진행
# (0<data['Role_Company'])  & (data['Role_Company'] < 0.4)
data['Role_Company_gb'] = np.where(data['Role_Company'] < 0.4, '0.4미만',
                                 np.where(data['Role_Company'] < 0.7, '0.4-0.7미만', '0.7이상'))
data[['Role_Company', 'Role_Company_gb']]
```

```python
# 근속연수 대비 같은 일을 한 데이터의 이직률 확인
gb = data.groupby('Role_Company_gb')['Attrition'].agg(['count', 'sum'])
gb['ratio'] = round(gb['sum'] / gb['count'] * 100, 1)
gb
# 결과 : 근속 년수 대비 같은 일을 짧게 할수록 이직률이 높다
```

- 가설 3) 야근을 많이 해도 급여 인상률이 높음면 이직률이 낮을 것이다!

```python
# 야근 여부에 따른 이직률 현황
gb = data.groupby('OverTime')['Attrition'].agg(['count', 'sum'])
gb['ratio'] = round(gb['sum'] / gb['count'] * 100, 1)
gb
# 결과 : 야근을 안 하는 사람보다 야근을 하는 사람이 이직률이 높다
```

```python
# 연봉 인상률 추가 (PercentSalaryHike)
gb = data.groupby(['OverTime', 'PercentSalaryHike'])['Attrition'].agg(['count', 'sum'])
gb['ratio'] = round(gb['sum'] / gb['count'] * 100, 1)
gb
```

```python
# 야근 하는 직원들의 연봉 인상률에 따른 이직률 확인
reset_gb = gb.reset_index()
# 야근을 하는 직원들만 추출
yes_gb = reset_gb[reset_gb['OverTime'] == 'Yes']
yes_gb
```

```python
# 연봉 인상률(x) 대비 이직률(y) 차트 그리기
plt.bar(yes_gb['PercentSalaryHike'],
        yes_gb['ratio'])
# 결과 : 야근을 많이 하는 직원은 연봉 인상률이 높음에도 불구하고 이직률이 높은 것을 확인
# -> 야근을 줄여주는 조치를 취해야겠다!
```

### 모델링
- 데이터 분리
- 모델 학습
- 모델 평가

```python
# 문제(X)데이터, 정답(y)데이터로 분리
data.info()
# 필요한 컬럼만 추출 -> 불필요한 컬럼은 삭제
X = data.drop(['Role_Company', 'Role_Company_gb', 'Attrition'], axis = 1)
y = data['Attrition']
```

```python
X.info()
# 데이터 결측치, 데이터 타입 확인
# 왜냐하면 문자열 데이터 타입은 학습이 불가능 -> 인코딩 작업 필요
# 지금까지 한 인코딩 작업 : 워낫인코딩
# 우선순위를 주는 인코딩 작업이 하고 싶다면 : 레이블 인코딩
```

#### 인코딩
- 범주형 데이터를 수치형 데이터로 변환
  - 원핫인코딩(onehot encoding)
    - 외부에 있는 데이터 값의 크고 작음의 의미가 없을 때
  - 레이블인코딩(label encoding)
    - 데이터 값의 크고 작음에 의미가 있어 가중치를 주고 싶을 때 -> 의미가 있을수록 큰 숫자 매칭

```python
X['BusinessTravel'].unique()
# 출장을 많이 갈수록 이직률이 높을 것이다!
```

```python
# 딕셔너리로 우선순위를 정의 (가중치를 주고 싶은 데이터에 큰 숫자를 매칭)
BT_dict = {'Non-Travel' : 1,
           'Travel_Rarely' : 2,
           'Travel_Frequently' : 3}
X['BusinessTravel'] = X['BusinessTravel'].map(BT_dict)
# map() : 키 값으로 검색하여 밸류로 변경해주는 함수
```

```python
# 원핫인코딩
X_one_hot = pd.get_dummies(X, dtype = 'int64')
```

```python
X_one_hot.info()
```

```python
# train, test 데이터로 분리
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_one_hot, y, test_size = 0.3, random_state = 1)
```

```python
# 데이터 크기 확인
print('훈련용 문제 : ', X_train.shape)
print('훈련용 정답 : ', y_train.shape)

print('테스트용 문제 : ', X_test.shape)
print('테스트용 정답 : ', y_test.shape)
```

- LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
# 이름에 Regression이 들어가지만 분류 모델이다!!
logi_model = LogisticRegression(max_iter = 10000) # max_iter 높을수록 정확도는 올라가지만, 과적합이 일어날 수 있어 -> 교차검증이 필요해
logi_model.fit(X_train, y_train)
logi_model.score(X_test, y_test)
```

### 교차검증
- 모델의 일반화 성능을 파악하자!
  - 학습-평가 데이터를 골고루 설정하여 모델의 안정성을 높이고 과대 적합을 감소시키는 통계적 기법
  - 방법 : (훈련용= 학습용 데이터 내에서) 훈련 세트와 테스트 세트로 여러겹 나누어 평가
  - 교차 검증 결과가 평균적으로 비슷한 결과가 출력된다면 일반화된 모델이라고 평가
  - 한정적인 데이터 내에서 여러겹으로 나눠 여러번 평가하면 신뢰도가 올라감
  - 확인용!!

```python
from sklearn.model_selection import cross_val_score # 교차 검증 도구
# cross_val_score(모델명, 학습용 문제, 학습용 정답, cv = 교차검증의 횟수 (주로 5))
result = cross_val_score(logi_model, X_train, y_train, cv = 5)
print(result)
print(result.mean())
```

- SVM 모델

```python
from sklearn.svm import LinearSVC # support vector의 classification
svm_model = LinearSVC()
svm_model.fit(X_train, y_train)
svm_model.score(X_test, y_test) # 정확도 accuracy
# 정확도 : 전체 데이터에서 정답을 맞춘 데이터의 비율
```

##### 평가지표 확인
- 정확도 (accuracy)
  - 전체 데이터에서 정답을 맞춘 비율
- 재현율 (recall)
  - 실제 양성에서 예측 양성의 비율
- 정밀도 (precision)
  - 예측 양성에서 실제 양성의 비율
- F1 - score
  - 재현율과 정밀도의 조화평균으로 두개의 값이 조화로울수록 수치가 올라간다!

```python
from sklearn.metrics import classification_report # 평가지표 모음집
# classification_report(실제정답, 예측정답)

# 예측하기
pre = logi_model.predict(X_test)

# 비교하기
print(classification_report(y_test, pre))
# 결과 : 우리가 쓰기에는 좀 더 데이터 조정이 필요하다
```

##### 이직에 영향을 미치는 입력특성 확인

```python
coef = logi_model.coef_[0] # 가중치를 출력
pd.Series(coef, index = X_train.columns).sort_values(ascending = False)
# 결과 : 야근을 많이 하면 -> 이직률이 높다
# 결과 : 출장을 많이 하면 -> 이직률이 높다
```
